```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
clean_wine="Resources/clean_winemag_data.csv"
apclean_wine_df=pd.read_csv(clean_wine)
```

```python
apclean_wine_df.head()
```

```python
# The correlation coefficient between "Country" and "Price"


country_price_df = apclean_wine_df[['Bottle Name','Country','Rating','Price']]
sorted_df = country_price_df.sort_values('Price', ascending=False)
sorted_df.head()
```

```python
# The Pearson correlation coefficient between "Rating" and "Price"

rating = apclean_wine_df.iloc[:,9]
price = apclean_wine_df.iloc[:,8]
correlation = st.pearsonr(rating,price)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
#Best wine in each price bracket
wine_prices_bin=[0, 100, 500, 1000, 5000]
group_names = ['low', 'normal','medium','high']

sorted_df['Price Group'] = pd.cut(sorted_df['Price'],wine_prices_bin,labels=group_names, include_lowest=True)
sorted_df
```
