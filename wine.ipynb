```python
# Dependencies
import pandas as pd

# Load file from Resources
wine_file = "Resources/winemag-data-130k-v2.csv"

# Create dataframe
wine_df =pd.read_csv(wine_file)

wine_df.head()
```

```python
# List of columns
wine_df.columns
```

```python
wine_df.count()
```

```python
# Remove nulls from data
reduced_wine_df = wine_df.dropna(subset=['price'])
reduced_wine_df.count()
```

```python
reduced_wine_df = reduced_wine_df.dropna(subset=['country'])
reduced_wine_df.count()
```

```python
reduced_wine_df = reduced_wine_df.dropna(subset=['province'])
reduced_wine_df.count()
```

```python
reduced_wine_df = reduced_wine_df.dropna(subset=['variety'])
reduced_wine_df.count()
```

```python
#Remove columns from dataframe
smaller_reduced_wine_df = reduced_wine_df.loc[:,['country', 'description', 'designation', 'points','price', 'province', 'region_1','taster_name','taster_twitter_handle', 'title', 'variety', 'winery']]

smaller_reduced_wine_df.head()
```

```python
smaller_reduced_wine_df=smaller_reduced_wine_df.rename(columns={'country':'Country', 'description':'Description', 'designation':'Designation', 'points':'Rating','price':'Price', 'province':'Province', 'region_1':'Region','taster_name':'Sommelier','taster_twitter_handle':'Sommelier Twitter Handle', 'title':'Bottle Name', 'variety':'Type', 'winery':'Winery'})

smaller_reduced_wine_df.head()
```

```python
cleaned_wine_df=smaller_reduced_wine_df[['Bottle Name','Type','Winery','Designation','Region','Province','Country','Price','Rating','Sommelier','Sommelier Twitter Handle','Description']]
cleaned_wine_df.head()
```

```python
cleaned_wine_df.dtypes
```

```python
cleaned_wine_df['Rating']=cleaned_wine_df['Rating'].astype(float)
cleaned_wine_df.dtypes
```

```python
cleaned_wine_df.count()
```

```python
cleaned_wine_df[['Designation','Region','Sommelier','Sommelier Twitter Handle']]=cleaned_wine_df[['Designation','Region','Sommelier','Sommelier Twitter Handle']].fillna("N/A")
cleaned_wine_df.count()
```
